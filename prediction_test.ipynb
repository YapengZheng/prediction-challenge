```python
# Data manipulation, visualization
import pandas as pd
import numpy as np
import seaborn as sns
import time
import matplotlib.pyplot as plt

# package to standardize/normalize the data
import sklearn.preprocessing as pre

# we may not need these generic functions, just code up on our own
# to be more flexible 
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV

# package to calculate (estimated) predictive accuracy
from sklearn.metrics import mean_squared_error, make_scorer

# regularized linear regression model
from sklearn.linear_model import LassoCV, ElasticNetCV, RidgeCV, ElasticNet

# random forest
from sklearn.ensemble import RandomForestRegressor

# trees
from sklearn.tree import DecisionTreeRegressor

# deep neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
```

```python
# import the data, meps_train.csv and load it into a pandas dataframe
# the data is in the same directory as the notebook
data_train = pd.read_csv('meps_train.csv')
data_test = pd.read_csv('meps_holdout.csv')
```

```python
# dimension of the data
print(data_train.shape)
# Form some summary statistics
print(data_train.describe())
# drop variables that names contained 'ID'
data_train = data_train.drop(data_train.filter(like='ID').columns, axis=1)
data_test = data_test.drop(data_test.filter(like='ID').columns, axis=1)
```

```python
# use cross-validation lasso to train model
# define the model
modelLasso = LassoCV(cv=5, random_state=0)
# define the features and target variable, drop first 13 columns
X = data_train.drop('exptot', axis=1)
X = X.select_dtypes(include=[np.number])
X = X.dropna(axis=1)
x = X.drop(X.columns[:13], axis=1)
x = pre.scale(x)
y = data_train['exptot']
modelLasso.fit(x, y)
# print the MSE, us K-fold cross-validation
print(mean_squared_error(y, modelLasso.predict(x)))
```

```python
# use random tree to train model, use cross-validation
# define the model
modelTree = DecisionTreeRegressor()
# fit the model
modelTree.fit(x, y)
```

```python
# print the tree model
dot_data = export_graphviz(modelTree, out_file=None,  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
